```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

#notebook
```

```python
np.set_printoptions(formatter={'float': lambda x: "{0:0.2f}".format(x)})
t = np.arange(0.01, 10*np.pi, .01)
x = t*np.sin(t)
print("t =",t)
print("x =",x)
plt.plot(x)
```

```python
tau = 160
plt.plot(x[:-2*tau], x[tau:-tau])
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
tau = 160
ax.plot(x[:-int(1.5*tau)], x[tau:-int(0.5*tau)], zs=x[int(1.5*tau):])
```
