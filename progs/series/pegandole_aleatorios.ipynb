```python
import numpy as np
import pandas as pd

from bokeh.plotting import show, figure, output_notebook

output_notebook()
```

```python
df = pd.read_csv("AirPassengers.csv", index_col=0)
st = df["#Passengers"]
serie2 = st[-10:].copy()  # colita de st
serie1 = st[:-10].copy()  # st sin colita
serie3 = serie1.copy()    # supongamos que serie3 es la ESTACIONARIA

aleatorios = np.random.rand(10)
aaleatorios = np.cumsum(aleatorios)

#pegarle aaleatorios a serie3
st_aaleatorios = pd.Series(aaleatorios, index=pd.DatetimeIndex(start="1960-3", end="1961-1", freq="1m"))
serie4 = serie3.append(st_aaleatorios)

p=figure()
p.line(serie4.index.values, serie4.values)
show(p)
```
