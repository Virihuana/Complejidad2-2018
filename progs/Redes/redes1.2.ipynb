```python
import numpy as np
import networkx as nx
from matplotlib.pyplot import show

#from bokeh.plotting import show, figure, output_notebook

#output_notebook()
```

```python
def cadena(n):
    m = np.zeros((n,n))
    for i in range(n-1):
        m[i, i+1] = 1
        m[i+1, i] = 1
    return m

def ciclo(n):
    m = cadena(n)
    m[0, n-1] = 1
    m[n-1, 0] = 1
    return m

def estrella(n):
    m = np.zeros((n,n))
    for i in range(1, n):
        m[0, i] = 1
        m[i, 0] = 1
    return m

def completa(n):
    return np.ones((n,n))

def muestra(matriz, etiqueta):
    print(etiqueta)
    print(matriz)
    nx.draw(nx.from_numpy_matrix(matriz))
    show()
    
def bipartita_completa(a, b):
    "bipartita completa con el 1er subconjunto de tamaño a y el 2o de tamaño b, a+b = n"
    n = a+b
    m = np.zeros((n,n))
    m[0:a, a:n] = 1
    m[a:n, 0:a] = 1
    return m

def arbol_binario(n):
    m=np.zeros((n,n))
    def j(i):
        return i-1
    i=0
    for i in range(n//2):
        m[2*(i+1)-1, i] = 1
        m[i, 2*(i+1)-1] = 1
        if not (i==n//2-1 and n%2==0):
            m[2*(i+1), i] = 1
            m[i, 2*(i+1)] = 1
    return m
```
