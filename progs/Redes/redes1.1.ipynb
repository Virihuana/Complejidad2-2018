```python
import networkx as nx
```

```python
g = nx.DiGraph()
g.add_node(1)
g.add_node(2)
g.add_node(3)
g.add_node(4)
g.add_edge(1,2)
g.add_edge(2,3)
g.add_edge(3,4)
g.add_edge(4,1)
g.nodes()
```

```python
g.edges()
```

```python
for e in g.edges():
    print(e)
```

```python
nx.draw(g)
```

```python
pos = nx.spring_layout(g)
nx.draw_networkx_edges(g, pos)
nx.draw_networkx_nodes(g, pos)
import matplotlib.pyplot as plt
plt.show()
```
