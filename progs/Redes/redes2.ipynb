El **grado** de un nodo $v$, denotado $\delta(v)$ es el número de conecciones de $v$. Cuando la gráfica es dirigida se puede hablar del grado de salida $\delta_{out}(v)$ y el grado de entrada $\delta_{in}(v)$ denotando el número de aristas que salen y llegan a $v$ respectivamente.

Algunas gráficas tienen el mismo grado en todos sus nodos. Por ejemplo, un ciclo tiene grado 2, un cubo tiene grado 3, un toroide 4, etc. se suelen llamar gráficas *cuadráticas* a las de grado 2, *cúbicas* a las de 3 etc.

Ej. Petersen (cúbica)

La **distribución de grados** resulta de graficar los grados sobre el eje $x$ y el número de nodos con tales grados en el eje $y$.

La máxima distancia entre dos nodos de una gráfica se suele llamar **diámetro**.

La distancia máxima de un nodo $v$ hasta todos los demás es su **excentricidad**. (si la gráfica es disconexa, su excentricidad es $\infty$).

Los nodos que alcanzan la mínima excentricidad constituyen el **centro** de la gráfica y dicha excentricidad mínima se le llama **radio**



La **máxima isla** es la componente conexa con más nodos.

Un **clan** es una subgráfica completa de la gráfica original. Un problema importante es el localizar el clan más grande en una gráfica o su distribución de tamaños de clanes.

En general una isla dá idea de que tanto alcance tiene alguna relación y el clan dice que tan fuerte es.

El **coeficiente de conectividad** de un nodo $v$, dice que tanto los vecinos de $v$ son vecinos entre si.

O sea, si $e(v)$ es el número de aristas entre vecinos de $v$, y 

$ne(v)=\frac{\delta(v)(\delta(v)+1)}{2}$

es el número máximo de estas aristas, entonces

$cc(v)=e(v)/ne(v)$

```python
import networkx as nx

g = nx.Graph()
g.add_edges_from([(0,1),(0,2),(1,3),(1,4),(2,5),(2,6),(3,7),(3,8),(4,9), (4,10),
                  (5,11),(5,12), (6,13),(6,14),(7,15),(7,16)])
nx.draw(g)
d = g.degree()
d = dict(d).values()
d = list(d)

import numpy as np

from bokeh.plotting import show,figure,output_notebook
output_notebook()

f = figure(plot_width=300, plot_height=300)
h, x = np.histogram(d, bins=3)

f.quad(bottom=0, left=x[:-1], right=x[1:], top=h, color=["pink","red","yellow"], line_color="#3A5785")
f.line(x[:-1]+1/3, h)
show(f)
#plt.hist(d, bins=3)  con matplotlib
```
