```python
inicial = 0
x = inicial
y = [x]
for pasos in range(100):
    r = random.randint(0, 1)*2-1
    x = x + r
    y.append(x)

import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

NCAMINANTES = 10
NPASOS = 100
y = np.random.randint(0, 2, (NCAMINANTES, NPASOS))*2-1
y = np.cumsum(y, axis=0)
x = np.arange(NPASOS)
#len(y[3]), len(x)

for i in range(10):
    plt.plot(x, y[i])
```
